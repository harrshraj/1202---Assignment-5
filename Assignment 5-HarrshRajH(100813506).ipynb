```python
# Loading the required libraries
import glob
import numpy 
import matplotlib
import pandas as pd
import matplotlib.pyplot 
import matplotlib.pyplot as plt
import numpy as np
```

```python
def visualize(filename):
    """Plotting the line graphs"""
    data = np.genfromtxt(filename ,delimiter=',') # importing the datasets
    avg1 = np.around(data.mean(axis=0),decimals=2) # calculating the average
    max1 = np.max(data,axis=0) # calculating the maximum
    min1 = np.min(data,axis=0) # calculating the minimum
    
    # Plotting the graph
    fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize=(15,3))
    ax1.plot(avg1)
    ax2.plot(max1)
    ax3.plot(min1)
    
    # Labelling the graph
    ax1.set_ylabel('Average')
    ax2.set_ylabel('Max')
    ax3.set_ylabel('Min')
    ax1.set_xlabel('Feature#')
    ax2.set_xlabel('Feature#')
    ax3.set_xlabel('Feature#')
    
    fig.tight_layout()
    plt.show()
```

```python
def identify_issues(filename):
    """Identifying issues"""
    data = np.genfromtxt(filename ,delimiter=',') # importing the datasets
    if np.max(data[:,[0]])==0 and np.max(data[:,[20]])==20: #checking for the first condition
        print('I doubt the integrity of this dataset')
    elif np.sum(np.min(data,axis=0))==0: # checking for the second condition
        print('Why is this happening?')
    else:
        print('Dataset looks good')
```

```python
# Final output
dataset = sorted(glob.glob('data*.csv'))

for datasets in dataset[:4]:
    print("\n")
    print(datasets)
    visualize(datasets)
    identify_issues(datasets)
```
